```python
import pandas as pd
import os
import matplotlib.pylab as plt
%matplotlib inline
```

```python
df=pd.read_csv("TKHALLI.csv")
df.head()
```

```python
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])

# Set 'date' as the index
df.set_index('date', inplace=True)

# (Optional) Drop year, month, and day columns if you no longer need them
# df.drop(['year', 'month', 'day'], axis=1, inplace=True)

# Display the DataFrame
df
```

```python
start_date = '1980-01-01'
end_date = '1985-03-31'

# Filter the DataFrame
filtered_df = df.loc[start_date:end_date]
```

```python
average_prcp = filtered_df['prcp(mm/day)'].mean()
print("Average precipitation from Jan 1985 to Mar 2010:", average_prcp, "mm/day")
```

```python

# Line plot for the filtered period
filtered_df['prcp(mm/day)'].plot(figsize=(10, 4), title='Precipitation (mm/day)', ylabel='mm/day')
plt.grid(True)
plt.show()
```

```python
monthly_avg = df['prcp(mm/day)'].resample('ME').mean()

monthly_avg.plot(kind='bar', figsize=(16, 8), title='Monthly Average Precipitation')
plt.ylabel('mm/day')
plt.grid(True)
plt.show()
```
